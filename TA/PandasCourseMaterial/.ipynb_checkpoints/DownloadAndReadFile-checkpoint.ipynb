## Downloading and Reading Files

Below is the step by step guide to downloading and reading in a file using Pandas.

1. Download the required file from Canvas to your desktop
    * The files are in the *Files* panel in Canvas. 
        - Within files section there is a *Pandas* Folder that houses all required files
    
    * After downloading, note the directory where you have placed the file, for windows the default download location might be something like
    C:/Users/YOUR_USER_NAME/Downloads
    
    
2. After having downloaded the required file from Canvas to your desktop, we will use the Jupyter Magic Commands to change directory to the folder where you have placed the file
    * Magic commands or magic functions are enhancements that IPython offers compared to the standard Python shell. Magic commands act as convenient functions where Python syntax is not the most natural one. They are useful to embed invalid python syntax in their work flow.
    
    
3. We will use the "*%cd DIR*" magic command.
    * This line magic changes the current directory to the directory specified
    
    
4. We create a variable *LOCATION* and store the folder name in the variable, then pass it to the "%cd" magic command

```python
LOCATION = "C:/Users/Abhijeet/Documents/GitHub/TA/TA/PandasCourseMaterial"
%cd $LOCATION
```

5. To verify that you have actually changed the directory you can use another magic function "%ls" - which lists the content of the current directory

```python
%ls
```

6. Now suppose we wanted to load the *HouseData.csv" file that is in this folder. We can simply use the pandas.read_csv function as follows

```python
import pandas as pd
pd.read_csv('HouseData.csv')
```
